As equações para $x(p)$ e $t(p)$ são:

$$ x(p) = 2p\sum_i \frac{\Delta z_i}{(u_i^2-p^2)^{1/2}}$$

Para $u_i>p$.

$$t(p)=2\sum_i\frac{u_i^2\Delta z_i}{(u_i^2-p^2)^{1/2}}$$

Também para $u_i>p$.

Considerando $p=0.15s/km$:


$$x(0.15) = 2 \times 0.15 \times \left( \frac{3}{(0.25^2-0.15^2)^{1/2}} + \frac{3}{(0.167^2-0.15^2)^{1/2}} \right) = 16.76km$$

$$t(0.15) = 2 \times \left( \frac{0.25^2 \times 3}{(0.25^2-0.15^2)^{1/2}} + \frac{0.167^2 \times 3}{(0.167^2-0.15^2)^{1/2}} \right) = 4.15s$$

```python
import math

layers = [0.250,0.167,0.125]
p = 0.15
z=3
x=0
t=0

for u in layers:
        
    if(u<p):
        break
    
    denominador = math.sqrt(u*u-p*p)
    
    x = x+(2*p*(z/denominador))
    
    t = t+(2*((u*u*z)/denominador))

print('x('+str(0.15)+')='+str(x))
print('t('+str(0.15)+')='+str(t))
```
