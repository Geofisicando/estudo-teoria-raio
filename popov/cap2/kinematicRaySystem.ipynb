# O SISTEMA CINEMÁTICO DO RAIO E O TRAÇAMENTO DE RAIOS

A equação iconal $(\nabla \tau)^2 = 1/v^2(x_i)$ é uma equação diferencial parcial não linear
de primeira ordem. Estas equações são usualmente solucionadas para $\tau$
em termos do _método das características_ (BLEISTEIN, 1984). As características da equação iconal
são trajetórias, descritas por um sistema de equações diferenciais que podem ser
solucionadas facilmente a partir de métodos numéricos padrão (CERVENY, 2001).

Na teoria sísmica do raio, nós definimos os raios como características da equação iconal e
chamamos o sistema de equações diferenciais ordinárias para as características do sistema
de traçamento do raio ou o sistema de equações do raio. Em um meio com camadas,
nós precisamos suplementar a definição dos raios, baseados
nas características da equação iconal, com a lei de Snell nestes pontos onde o raio
entra em contato com as interfaces.

As _características da equação iconal_ são trajetórias espaciais 3D $x_i=x_i(u)$ ($u$ sendo um parâmetro monotônico ao longo
da trajetória do raio).
As curvas características são uma solução do chamado
sistema característico de equações diferenciais ordinárias de primeira ordem.

### O sistema cinemático do raio a partir do método das características

Partindo da equação iconal (RUEGER, 1993):

$$|\nabla\tau|^2=\frac{1}{v^2}$$

Esta pode
ser reescrita como,

$$\frac{1}{n}\left(|\nabla\tau\cdot\nabla\tau|^{n/2}-\frac{1}{v^n}\right)=0$$

O vetor $p_i$ (slowness),

$$p_i\equiv\nabla\tau$$

$$|p|=\frac{1}{v}$$

A equação iconal toma a forma,

$$H(p_i,x_i)=\frac{1}{n}\left[(p_ip_i)^{n/2}-\left(\frac{1}{v^n}\right)\right]=0$$

Esta equação é uma equação diferencial parcial não-linear de primeira ordem do tipo Hamilton-Jacobi.
Ou seja, ao longo da trajetória $H(x_i,p_i)=0$ é satisfeita. E utilizando o método das características,
nós obtemos as seguintes equações para o traçamento
de raio (BLEISTEIN, 1984):

$$du=v^nd\tau$$

$$\frac{dx_i}{du}=p_i\frac{1}{v^{n-2}}$$

$$\frac{dp_i}{du}=-\frac{1}{v^{n+1}}\frac{\partial v}{\partial x_i}$$

A solução destas equações representa a trajetória $x_i$ e a distribuição de $p_i$ ao longo do raio como uma
função de uma variável monotonicamente crescente e independente $u$. Estas soluções de $x_i(u)$ são chamadas
de raios. Somos livres para escolher $n$. Assim:

Para $n=0$, $u$ é o tempo de trânsito $\tau$ ao longo do raio.

$$du=d\tau$$

Para $n=1$, $u$ é o comprimento de arco $s$ ao longo do raio.

$$du=vd\tau=ds$$

Para $n=2$, $u$ é um parâmetro arbitrário $\sigma$.

$$du=v^2d\tau=d\sigma$$

Assim, para $u=\sigma$,

$$\frac{dx_i}{d\sigma}=p_i$$

$$\frac{dp_i}{d\sigma}=\frac{1}{2}\nabla \left(\frac{1}{v^2}\right)$$

$$\frac{d\tau}{d\sigma}=\frac{1}{v^2}$$

__Observe:__

$$\frac{dp_i}{d\sigma}=\frac{1}{2}\nabla \left(\frac{1}{v^2}\right)=
\frac{1}{2}-2 v^{-3}\frac{\partial v}{\partial x_i}
=-\frac{1}{v^3}\frac{\partial v}{\partial x_i}$$

### Reparametrização para o tempo de trânsito

Deste modo, se,

$$\frac{d\sigma}{d\tau}=v^2$$

Então, se $p^2=1/v^2$

$$\frac{dx_i}{d\sigma}\frac{d\sigma}{d\tau}=p_iv^2=\frac{p_i}{|p|^2}$$

$$\frac{dp_i}{d\sigma}\frac{d\sigma}{d\tau}=\frac{1}{2}\nabla \left(\frac{1}{v^2}\right)v^2=\frac{1}{2} \left(\frac{\nabla p^2}{|p|^2}\right)$$

O traçamento de raios é realizado a partir da integração numérica do sistema cinemático do raio
utilizando o método Runge-Kutta. O sistema é apresentado
a seguir (RUEGER, 1993):

\begin{equation}
\label{cine:eq:1.1}
\frac{dx_i}{d\tau}=p_iv^2
\end{equation}

\begin{equation}
\label{cine:eq:1.2}
\frac{dp_i}{d\tau}=\frac{1}{2}\nabla \left(\frac{1}{v^2}\right)v^2
\end{equation}

Para o caso 2D, temos,

\begin{equation}
\label{cine:eq:1.3}
i=1,2
\end{equation}

Onde $x_1$ representa a coordenada da profundidade e $x_2$ a coordenada em relação à horizontal.
Assim, $p_1$ é o vetor vagarosidade em profundidade e $p_2$ é o vetor vagarosidade na horizontal.
Deste modo, $v$ é o campo de velocidades em subsuperfície e $|p|=1/v$.

A partir das condições iniciais para a coordenada inicial do raio, $x_1$ e $x_2$ e para direção inicial, $p_1$ e $p_2$,
a trajetória do raio é obtida a partir das Equações do sistema cinemático do raio por integração numérica.
